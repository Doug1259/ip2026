Apply the Laplacian filtering to the downloaded grayscale image, then:
- Display the original image, the Laplacian image and the sharpened image.
- Compare and explain the results before and after sharpening.

### The original image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("grayscaledcoolpicture.jpeg", 0)

plt.imshow(image, cmap="gray")
plt.axis("off")
plt.title("grayscaledcoolpicture.jpeg")
plt.show()
```

### The Laplacian filter

```python
laplacian_kernel=np.array([
    [0, 1, 0],
    [1, -4, 1],
    [0, 1, 0]
])

def laplacian_filter(image):
    output = np.zeros((image.shape[0],image.shape[1]))
    for i in range(1, image.shape[0] - 1):
        for j in range(1, image.shape[1] - 1):
            window=image[i-1:i+2, j-1:j+2]
            output[i, j]=(window*laplacian_kernel).sum()
    return output
```

```python
laplacian_image=laplacian_filter(image)

plt.imshow(laplacian_image,cmap="gray")
plt.axis("off")
plt.title("The laplacian image")
plt.show()
```

### The sharpened image

```python
sharpened_image= image - laplacian_image

plt.imshow(sharpened_image, cmap="gray")
plt.axis("off")
plt.title("The sharpened image")
plt.show()
```

```python
comp = np.hstack((image,sharpened_image))

plt.imshow(comp, cmap="gray")
plt.axis("off")
plt.title("Original vs sharpened")
plt.show()
```

The image got sharper

The laplacian filter first detect the place where the pixel values next to each other change drastically, and return an image containing only the changes at their position. Then it takes the original then subtract the image with the edges only which adds more value to the original pixels bordering on the edges, resulting in a sharper image
