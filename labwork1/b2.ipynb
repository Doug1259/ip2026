# Task B2: Brightness and contrast adjustment
&emsp;Change the brightness and contrast of the downloaded image using the following transformation:

&emsp;&emsp;&emsp;&emsp;&emsp;$img_{processed}$ = a * f(x,y) + b

&emsp;In which, 𝑓(𝑥,𝑦) is the pixel value of the original image (𝑖𝑚𝑔) at the coordinate (𝑥, 𝑦); a is a user-defined contrast control parameter; b is a user-defined brightness control parameter. Apply the four following different pairs of (a,b):

&emsp;&emsp;&emsp;&emsp;&emsp;(a,b) = (1.0,50) or (1.0,−50) or (1.5,0) or (0.5,0)

&emsp;Then, complete the following tasks:
- display the original and processed images;
- display the histogram of each processed image;
- compare the effects of different values of a and b;
- explain how a affects contrast and how b affects brightness

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

Brightness and contrast adjustment function:

```python
def bricon(image, a, b):
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            pixel=int(image[i, j][0])
            processed=a*pixel+b
            if processed<0:
                processed=0
            elif processed>255:
                processed=255
            image[i,j]=processed
```

The original image

```python
grayimage = cv2.imread("grayscaleddropleaf.png")

plt.imshow(grayimage)
plt.axis("off")
plt.title("Original image")
plt.show()
```

Processed image with (a,b)=(1,50)

```python
processedimage1 = grayimage.copy()
bricon(processedimage1,1,50)

fig, axes = plt.subplots(
    1, 2,
    figsize=(12, 4),
    width_ratios=[1, 2]
)

# Image
axes[0].imshow(processedimage1)
axes[0].axis("off")
axes[0].set_title("Processed image with (a,b)=(1,50)")

# Histogram
axes[1].hist(processedimage1.ravel(), bins=256, range=[0,256])
axes[1].set_title("Histogram of the processed image 1")
axes[1].set_xlabel("Intensity")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

Processed image with (a,b)=(1,-50)

```python
processedimage2 = grayimage.copy()
bricon(processedimage2,1,-50)

fig, axes = plt.subplots(
    1, 2,
    figsize=(12, 4),
    width_ratios=[1, 2]
)

# Image
axes[0].imshow(processedimage2)
axes[0].axis("off")
axes[0].set_title("Processed image with (a,b)=(1,-50)")

# Histogram
axes[1].hist(processedimage2.ravel(), bins=256, range=[0,256])
axes[1].set_title("Histogram of the processed image 2")
axes[1].set_xlabel("Intensity")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

Processed image with (a,b)=(1.5,0)

```python
processedimage3 = grayimage.copy()
bricon(processedimage3,1.5,0)

fig, axes = plt.subplots(
    1, 2,
    figsize=(12, 4),
    width_ratios=[1, 2]
)

# Image
axes[0].imshow(processedimage3)
axes[0].axis("off")
axes[0].set_title("Processed image with (a,b)=(1.5,0)")

# Histogram
axes[1].hist(processedimage3.ravel(), bins=256, range=[0,256])
axes[1].set_title("Histogram of the processed image 3")
axes[1].set_xlabel("Intensity")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

Processed image with (a,b)=(0.5,0)

```python
processedimage4 = grayimage.copy()
bricon(processedimage4,0.5,0)

fig, axes = plt.subplots(
    1, 2,
    figsize=(12, 4),
    width_ratios=[1, 2]
)

# Image
axes[0].imshow(processedimage4)
axes[0].axis("off")
axes[0].set_title("Processed image with (a,b)=(0.5,0)")

# Histogram
axes[1].hist(processedimage4.ravel(), bins=256, range=[0,256])
axes[1].set_title("Histogram of the processed image 4")
axes[1].set_xlabel("Intensity")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

- b shifts the histogram to the left if it is negative or to the right if it is positive
- a "extends" the histogram if it is larger than 1 and "compress" it if it is smaller than 1

- b either makes all the pixels brighter or darker
- with a > 1, the light pixel values increase more substantially than the dark pixel values which create a large "distance" between dark and light, causing high contrast
- with a < 1, the light pixel values decrease and move nearer to the dark pixel values which create a small "distance" between dark and light, causing low contrast
