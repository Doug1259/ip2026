# Task B3: Image Thresholding
&emsp;Perform image thresholding to convert a grayscale image into a binary image.
- Read and display an image. If the selected image is a color image, convert it to grayscale first.
- Display the histogram of the grayscale image.
- Apply global thresholding using the three predefined following threshold values:

&emsp;&emsp;&emsp;&emsp;&emsp;T = 80 or 120 or 160

- Display the binary images obtained from different threshold values.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

Threshold function:

```python
def thresholding(image,threshold):
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            pixel=image[i, j][0]
            if pixel>threshold:
                pixel=255
            elif pixel<=threshold:
                pixel=0
            image[i, j]=pixel
```

The original image

```python
grayimage = cv2.imread("grayscaledcoolpicture.jpeg")

fig, axes = plt.subplots(
    1, 2,
    figsize=(12, 4),
    width_ratios=[1, 2]
)

# Image
axes[0].imshow(grayimage)
axes[0].axis("off")
axes[0].set_title("Gray image")

# Histogram
axes[1].hist(grayimage.ravel(), bins=256, range=[0,256])
axes[1].set_title("Histogram of the gray image")
axes[1].set_xlabel("Intensity")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

T=80

```python
processedimage1=grayimage.copy()
thresholding(processedimage1,80)

plt.imshow(processedimage1)
plt.axis("off")
plt.title("Threshold T = 80")
plt.show()
```

T=120

```python
processedimage1=grayimage.copy()
thresholding(processedimage1,120)

plt.imshow(processedimage1)
plt.axis("off")
plt.title("Threshold T = 120")
plt.show()
```

T=160

```python
processedimage1=grayimage.copy()
thresholding(processedimage1,160)

plt.imshow(processedimage1)
plt.axis("off")
plt.title("Threshold T = 160")
plt.show()
```
