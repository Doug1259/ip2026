Read the downloaded images using OpenCV. If an image is a color image, convert it to grayscale. Then perform the following tasks:
#### (a) Histogram and histogram equalization:
- Compute and display the histogram of the grayscale image.
- Apply histogram equalization to the grayscale image.
- Compute and display the histogram of the equalized image.
#### (b) Resize the grayscale image to 150% of its original width and height.
#### (c) Display the following results:
- original image;
- grayscale image;
- resized grayscale image;
- histogram of the grayscale image;
- equalized grayscale image;
- histogram of the equalized grayscale image;
- grayscale image and equalized grayscale image side by side for comparison.
#### (d) Print the basic information of both the original image and the grayscale image, including:
- image size;
- number of channels;
- data type;
- minimum pixel value;
- maximum pixel value.

### Reading the images

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image1 = cv2.imread("coolpicture.jpeg")
image2 = cv2.imread("dropleaf.png")
RGB_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
RGB_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

plt.imshow(RGB_image1)
plt.axis("off")
plt.title("coolpicture.jpeg")
plt.show()

plt.imshow(RGB_image2)
plt.axis("off")
plt.title("dropleaf.png")
plt.show()
```

### Converting them to grayscale

```python
def grayscale(image,name):
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            pixel=image[i, j]
            gray=int(pixel[0]/3)+int(pixel[1]/3)+int(pixel[2]/3)
            image[i, j]=gray
    cv2.imwrite(name,image)
```

```python
grayscale(RGB_image1.copy(),"grayscaledcoolpicture.jpeg")
grayscale(RGB_image2.copy(),"grayscaleddropleaf.png")

grayimage1 = cv2.imread("grayscaledcoolpicture.jpeg")
grayimage2 = cv2.imread("grayscaleddropleaf.png")

plt.imshow(grayimage1)
plt.axis("off")
plt.title("grayscaledcoolpicture.jpeg")
plt.show()

plt.imshow(grayimage2)
plt.axis("off")
plt.title("grayscaleddropleaf.png")
plt.show()
```

### (a) Histogram and histogram equalization:

#### Histograms of the grayscaled images

Histogram of grayscaled image 1

```python
fig, axes = plt.subplots(
    1, 2,
    figsize=(12, 4),
    width_ratios=[1, 2]
)

# Image
axes[0].imshow(grayimage1)
axes[0].axis("off")
axes[0].set_title("grayimage1")

# Histogram
axes[1].hist(grayimage1.ravel(), bins=256, range=[0,256])
axes[1].set_title("Histogram of grayimage1")
axes[1].set_xlabel("Intensity")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

Histogram of grayscaled image 2

```python
fig, axes = plt.subplots(
    1, 2,
    figsize=(12, 4),
    width_ratios=[1, 2]
)

# Image
axes[0].imshow(grayimage2)
axes[0].axis("off")
axes[0].set_title("grayimage2")

# Histogram
axes[1].hist(grayimage2.ravel(), bins=256, range=[0,256])
axes[1].set_title("Histogram of grayimage2")
axes[1].set_xlabel("Intensity")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

#### Applying histogram equalization

```python
actual_grayimage1= cv2.imread("grayscaledcoolpicture.jpeg",0)
actual_grayimage2 = cv2.imread("grayscaleddropleaf.png",0)

equimage1 = cv2.equalizeHist(actual_grayimage1)
equimage2 = cv2.equalizeHist(actual_grayimage2)
```

Histogram of equalized image 1

```python
fig, axes = plt.subplots(
    1, 2,
    figsize=(12, 4),
    width_ratios=[1, 2]
)

# Image
axes[0].imshow(equimage1,cmap="gray")
axes[0].axis("off")
axes[0].set_title("equimage1")

# Histogram
axes[1].hist(equimage1.ravel(), bins=256, range=[0,256])
axes[1].set_title("Histogram of equimage1")
axes[1].set_xlabel("Intensity")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

Histogram of equalized image 2

```python
fig, axes = plt.subplots(
    1, 2,
    figsize=(12, 4),
    width_ratios=[1, 2]
)

# Image
axes[0].imshow(equimage2,cmap="gray")
axes[0].axis("off")
axes[0].set_title("equimage2")

# Histogram
axes[1].hist(equimage2.ravel(), bins=256, range=[0,256])
axes[1].set_title("Histogram of equimage2")
axes[1].set_xlabel("Intensity")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

### (b) Resize the grayscale image to 150% of its original width and height.

```python
nheight1=int(grayimage1.shape[0]*1.5)
nwidth1=int(grayimage1.shape[1]*1.5)
nheight2=int(grayimage2.shape[0]*1.5)
nwidth2=int(grayimage2.shape[1]*1.5)
```

```python
rsgrayimage1=cv2.resize(grayimage1,(nwidth1,nheight1))
rsgrayimage2=cv2.resize(grayimage2,(nwidth2,nheight2))
```

### (c) Display the following results:

Original images

```python
fig, axes = plt.subplots(1, 2)

fig.suptitle("The originals")

axes[0].imshow(RGB_image1)
axes[0].axis("off")
axes[0].set_title("original 1")

axes[1].imshow(RGB_image2)
axes[1].axis("off")
axes[1].set_title("original 2")

plt.show()
```

Grayscale images

```python
fig, axes = plt.subplots(1, 2)

fig.suptitle("The grayscales")

axes[0].imshow(grayimage1)
axes[0].axis("off")
axes[0].set_title("grayscale 1")

axes[1].imshow(grayimage2)
axes[1].axis("off")
axes[1].set_title("grayscale 2")

plt.show()
```

Resized grayscale images

```python
fig, axes = plt.subplots(1, 2)

fig.suptitle("The resized grayscales")

axes[0].imshow(rsgrayimage1)
axes[0].set_xticks([])
axes[0].set_yticks([])
axes[0].set_title("resized grayscale 1")
axes[0].set_xlabel(f"Resized: {rsgrayimage1.shape[1]}x{rsgrayimage1.shape[0]} vs Original: {grayimage1.shape[1]}x{grayimage1.shape[0]}")

axes[1].imshow(rsgrayimage2)
axes[1].set_xticks([])
axes[1].set_yticks([])
axes[1].set_title("resized grayscale 2")
axes[1].set_xlabel(f"Resized: {rsgrayimage2.shape[1]}x{rsgrayimage2.shape[0]} vs Original: {grayimage2.shape[1]}x{grayimage2.shape[0]}")

plt.show()
```

Equalized grayscale images

```python
fig, axes = plt.subplots(1, 2)

fig.suptitle("The equalized grayscales")

axes[0].imshow(equimage1,cmap="gray")
axes[0].axis("off")
axes[0].set_title("equalized grayscale 1")

axes[1].imshow(equimage2,cmap="gray")
axes[1].axis("off")
axes[1].set_title("equalized grayscale 2")

plt.show()
```

Grayscale images vs Equalized grayscale images

```python
comp1=np.hstack((actual_grayimage1,equimage1))
comp2=np.hstack((actual_grayimage2,equimage2))

fig, axes = plt.subplots(
    1, 2,
    figsize=(12,4)
)

fig.suptitle("The grayscales vs equalized grayscale images")

axes[0].imshow(comp1,cmap="gray")
axes[0].axis("off")
axes[0].set_title("grayscale 1(L) vs equalized grayscale 1(R)")

axes[1].imshow(comp2,cmap="gray")
axes[1].axis("off")
axes[1].set_title("grayscale 2(L) vs equalized grayscale 2(R)")

plt.show()
```
